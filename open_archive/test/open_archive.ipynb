```python
import mlrun, os
mlrun.mlconf.dbpath="http://mlrun-api:8080"
```

```python
# ZIP_ARCHIVE = "http://iguazio-sample-data.s3.amazonaws.com/catsndogs.zip"
TAR_ARCHIVE = "https://fpsignals-public.s3.amazonaws.com/catsndogs.tar.gz" # no permission needs to be replaced with other data
```

```python
FUNCTION = "open_archive"

task_params = {
    "name":        FUNCTION, 
    "params"   : {
        "key"   : "tar-data",
        "archive_url" : TAR_ARCHIVE
    },
    "artifact_path"  : os.path.join(f"/User/functions/{FUNCTION}/artifacts", 
                                    "{{run.uid}}")
}
```

```python
rfn = mlrun.import_function(f"/User/functions/{FUNCTION}/function.yaml").apply(mlrun.mount_v3io())
tsk = rfn.run(mlrun.NewTask(**task_params))
```
