```python
import mlrun, os, json
mlrun.mlconf.dbpath = 'http://mlrun-api:8080'
```

```python
config_fn = mlrun.import_function(os.path.join(os.getcwd(), 'function.yaml')).apply(mlrun.mount_v3io())
```

```python
ARTIFACT_PATH = os.path.join(os.getcwd(), 'artifacts')
os.makedirs(ARTIFACT_PATH, exist_ok=True)
```

```python
config_task = mlrun.NewTask(
    'get-config',
    params={'config_url' : os.path.join(os.getcwd(), 'xgboost-conf.json')},
    artifact_path=ARTIFACT_PATH)

dc = config_fn.run(config_task)
```
