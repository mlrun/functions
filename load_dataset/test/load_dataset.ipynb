```python
import mlrun, os
mlrun.mlconf.dbpath="http://mlrun-api:8080"
```

```python
FUNCTION = "load_dataset"

task_params = {
    "name"           : FUNCTION, 
    "params"         : {
        "dataset"     : "breast_cancer"
    },
    "artifact_path"  : os.path.join(f"/User/functions/{FUNCTION}")
}
```

```python
rfn = mlrun.import_function(f"/User/functions/{FUNCTION}/function.yaml").apply(mlrun.mount_v3io())
tsk = rfn.run(mlrun.NewTask(**task_params))
```

```python
import pandas as pd
df = pd.read_parquet(tsk.outputs["breast_cancer"])
df
```
