# OAI Hub Demo


## Overview


This notebook demonstrates how to use the OAI Hub class for managing MLRun project setup and artifact logging.

The OaiHub class handles:
- Environment configuration
- Project creation/retrieval
- Input data processing and logging
- Configuration dataset logging


## Demo

```python
from oai_spo import OaiHub

# Initialize OaiHub instance
oai_hub = OaiHub(
    project_name="oai-spo-demo",
    data_dir="./data",
    default_env_file="default.env",
    local_env_file="local.env",
    pipeline_config_path="pipeline_config.yaml",
    default_image="mlrun/mlrun",
    source="s3://your-bucket/path",
)
```
