```python
import mlrun
mlrun.mlconf.mlrundb = 'https://mlrun-api:8080'
```

## test each url type

```python
urls = ['function.yaml',
        'https://raw.githubusercontent.com/yjb-ds/functions/init_fixes/describe/function.yaml',
        'db://describe',
        'hub://describe']
```

```python
for url in urls:
    rfn = mlrun.import_function(url).apply(mlrun.mount_v3io())
    tsk = rfn.run(
            mlrun.NewTask(
                "sum", 
                params={"key": "summary", "label_column": "species"},
                inputs={"table": '/User/functions/describe/iris.pqt'},
                artifact_path='/User/functions/describe/artifacts/{{run.uid}}'))
```
