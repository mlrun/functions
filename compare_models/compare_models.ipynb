```python
import os
import mlrun
```

```python
fn = mlrun.code_to_function(
    name="compare-models",
    filename="compare_models.py",
    kind="job",
    handler="compare_models",
    with_doc=False,
    image=".mlrun/func-default-compare-models:latest",
    #image="mlrun/mlrun",
    #requirements="requirements.txt"
).apply(mlrun.auto_mount())
```

```python
#fn.deploy(skip_deployed=True)
```

```python
fn.run(
    inputs={
        "reference_model_path" : f"{os.getcwd()}/lr.pkl",
        "current_model_path" : f"{os.getcwd()}/rf.pkl",
        "data_path" : f"{os.getcwd()}/iris.csv"
    },
    params={
        "target": "target",
        "class_mappings": {0 : 'setosa', 1 : 'versicolor', 2 : 'virginica'},
        "comparison_metric": "accuracy"
    }
)
```
